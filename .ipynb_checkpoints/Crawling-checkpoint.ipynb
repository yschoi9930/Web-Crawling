```python
years = ['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']
```

```python
type(year)
```

```python
final_oyster=pd.DataFrame()
for year in years :
    oyster = pd.read_csv("./data/굴_"+year+"/굴_"+year+".csv")
    final_oyster=pd.concat([final_oyster,oyster], axis=0)
    final_oyster.to_csv('./data/final/oyster.csv')
```

```python
final_abalone=pd.DataFrame()
for year in years :
    abalone = pd.read_csv("./data/전복_"+year+"/전복_"+year+".csv")
    final_abalone=pd.concat([final_abalone,abalone], axis=0)
    final_abalone.to_csv('./data/final/abalone.csv')
```

```python
final_mackerel=pd.DataFrame()
for year in years :
    mackerel = pd.read_csv("./data/삼치_"+year+"/삼치_"+year+".csv")
    final_mackerel=pd.concat([final_mackerel,mackerel], axis=0)
    final_mackerel.to_csv('./data/final/mackerel.csv')
```
